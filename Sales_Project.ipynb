```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dfe = pd.read_csv('/content/sales_data.csv', encoding="ISO-8859-1")
dfe
```

```python
dfe['Amount'].isnull().sum()
```

```python
dfe=dfe.drop(['Status', 'unnamed1'], axis=1)
dfe
```

```python
dfe.dropna(inplace=True)
dfe
```

```python
dfe.info()
```

```python
dfe.rename(columns={'Marital_Status' : 'Marital'}, inplace=True)
dfe
```

```python
dfe['Marital'] = dfe['Marital'].replace([0, 1], ['Single', 'Married'])
dfe
```

```python
dfe['Gender'].unique()
```

```python
dfe['Gender'] = dfe['Gender'].replace(['F', 'M'], ['Female', 'Male'])
dfe
```

```python
plt.figure(figsize=(10, 7), facecolor = 'lightblue')
plt.title("Count of Males and Females based on Gender", fontstyle='italic')
ax = plt.gca()
ax.set_facecolor('lavender')
a=sns.countplot(x='Gender', data=dfe, color='darkcyan', hue = 'Gender')

for i in a.containers:
  a.bar_label(i)
```

```python
dfe
```

```python
a=plt.figure(figsize=(60,40), facecolor='lightgrey')
plt.title("Amount Expenditure", fontstyle='oblique', fontsize=60, x=0.5,y=1.05)

plt.subplot(3,3,1)
ax = plt.gca()
ax.set_facecolor('lavender')
ax.set_title('Based on Gender', fontsize=35)
a=sns.barplot(x='Gender', y='Amount', data=dfe, palette='dark:darkcyan', edgecolor='black', errorbar=None)
for i in a.containers:
  a.bar_label(i)

plt.subplot(3,3,2)
ax = plt.gca()
ax.set_facecolor('lavender')
ax.set_title('Based on Age Group', fontsize=35)
a=sns.barplot(x='Age Group', y='Amount', data=dfe, palette='dark:darkcyan', edgecolor='black', errorbar=None)
for i in a.containers:
  a.bar_label(i)
plt.xticks(rotation=90)

plt.subplot(3,3,3)
ax = plt.gca()
ax.set_facecolor('lavender')
ax.set_title('Based on State', fontsize=35)
a=sns.barplot(x='State', y='Amount', data=dfe, palette='dark:darkcyan', edgecolor='black', errorbar=None)
for i in a.containers:
  a.bar_label(i)
plt.xticks(rotation=90)

plt.subplot(3,3,4)
ax = plt.gca()
ax.set_facecolor('lavender')
ax.set_title('Based on Occupation', fontsize=35)
a=sns.barplot(x='Occupation', y='Amount', data=dfe, palette='dark:darkcyan', edgecolor='black', errorbar=None)
for i in a.containers:
  a.bar_label(i)
plt.xticks(rotation=90)

plt.subplot(3,3,5)
ax = plt.gca()
ax.set_facecolor('lavender')
ax.set_title('Based on Product Category', fontsize=35)
a=sns.barplot(x='Product_Category', y='Amount', data=dfe, palette='dark:darkcyan', edgecolor='black', errorbar=None)
for i in a.containers:
  a.bar_label(i)
plt.xticks(rotation=90)

plt.subplot(3,3,6)
ax = plt.gca()
ax.set_facecolor('lavender')
ax.set_title('Based on Marital Status', fontsize=35)
a=sns.barplot(x='Marital', y='Amount', data=dfe, palette='dark:darkcyan', edgecolor='black', errorbar=None)
for i in a.containers:
  a.bar_label(i)

plt.show()
```

```python
plt.figure(figsize=(40,35), facecolor='lavender')


plt.subplot(3,3,1)
a=dfe.pivot_table(values=['Amount'], index='Gender', aggfunc=['mean','sum'])
sns.heatmap(a,annot=True, cmap='Blues', linewidths=2, linecolor='black', fmt='.2f')
ax=plt.gca()
ax.set_title('Based on Gender', fontsize=35)

plt.subplot(3,3,2)
a=dfe.pivot_table(values=['Amount'], index='Age Group', aggfunc=['mean','sum'])
sns.heatmap(a,annot=True, cmap='Blues', linewidths=2, linecolor='black', fmt='.2f')
ax = plt.gca()
ax.set_title('Based on Age Group', fontsize=35)

plt.subplot(3,3,3)
a=dfe.pivot_table(values=['Amount'], index='State', aggfunc=['mean','sum'])
sns.heatmap(a,annot=True, cmap='Blues', linewidths=2, linecolor='black', fmt='.2f')
ax = plt.gca()
ax.set_title('Based on State', fontsize=35)

plt.subplot(3,3,4)
a=dfe.pivot_table(values=['Amount'], index='Occupation', aggfunc=['mean','sum'])
sns.heatmap(a,annot=True, cmap='Blues', linewidths=2, linecolor='black', fmt='.2f')
ax = plt.gca()
ax.set_title('Based on Occupation', fontsize=35)

plt.subplot(3,3,5)
a=dfe.pivot_table(values=['Amount'], index='Product_Category', aggfunc=['mean','sum'])
sns.heatmap(a,annot=True, cmap='Blues', linewidths=2, linecolor='black', fmt='.2f')
ax = plt.gca()
ax.set_title('Based on Product Category', fontsize=35)

plt.subplot(3,3,6)
a=dfe.pivot_table(values=['Amount'], index='Marital', aggfunc=['mean','sum'])
sns.heatmap(a,annot=True, cmap='Blues', linewidths=2, linecolor='black', fmt='.2f')
ax = plt.gca()
ax.set_title('Based on Marital Status', fontsize=35)

plt.show()
```
